```python
from selenium import webdriver
import os
import time
from PIL import Image
from io import BytesIO
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

def acesso_power_bi(senhauser, emailuser, caminhoService, linkPowerBi, caminhoCaptura):

    service = Service(executable_path=caminhoService)
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=service, options=options)

    driver.get(linkPowerBi)

    time.sleep(1)

    time.sleep(6)

    email = driver.find_element(by=By.CLASS_NAME, value='pbi-text-input')

    time.sleep(5)

    email.send_keys(emailuser)
    time.sleep(5)

    submit = driver.find_element(by=By.ID, value='submitBtn')
    time.sleep(5)

    submit.click()
    time.sleep(5)

    senha = driver.find_element(by=By.ID, value='i0118')

    senha.send_keys(senhauser)
    time.sleep(5)

    submit2 = driver.find_element(by=By.ID, value='idSIButton9')
    time.sleep(5)
    submit2.click()
    time.sleep(15)

    screenshot = driver.get_screenshot_as_png()

    latitude = 100
    longitude = 200

    element_x = longitude
    element_y = latitude

    element_width = 800
    element_height = 600

    box = (element_x, element_y, element_x + element_width, element_y + element_height)
    element_screenshot = Image.open(BytesIO(screenshot)).crop(box)

    element_screenshot.save(caminhoCaptura)

    driver.quit()

    return caminhoCaptura

# Substitua os marcadores de posição pelos seus dados reais
senhauser = 'SUA_SENHA_AQUI'
emailuser = 'SEU_EMAIL_AQUI'
caminhoService = r'CAMINHO_PARA_O_CHROMEDRIVER'
linkPowerBi = r'LINK_DO_POWER_BI'
caminhoCaptura = r'C:\Caminho\Para\Salvar\Imagem.png'

# Chama a função e guarda o caminho da imagem capturada
image_path = acesso_power_bi(senhauser, emailuser, caminhoService, linkPowerBi, caminhoCaptura)


```
